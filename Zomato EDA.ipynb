```python
ratings=final_data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

When the ratings is between 4.5-4.9 -----> Excellent

When the ratings is between 4.0-4.4 ------> Very Good

When the ratings is between 3.5-3.9 -------> Good

When the ratings is between 2.5-3.4 -------> Average

When the ratings is between 1.8-2.4 -------> Poor 


```python
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```

1.No rated counting are very high

2.Maximum rating are between 2.5 to 3.4

```python
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
